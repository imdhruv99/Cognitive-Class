### Data Preprocessing [ Data Wrangling or Data Cleaning ]

#### **Missing Value**

- Normally in Dataset missing values can be appear as "?", "N/A", "0" or just blacnk cell.

```python
df.head()
```

- As you can see, "normalized-losses" contains missing values.

##### Methods To Remove Missing Values

    - Drop the Missing Values
        - Drop the Variable
        - Drop the Data Entry
    
    - Replace Missing Value
        - Replace it with an average
        - Replace it by Frequency
        - Replace it based on other functions

```python
# conver "?"  to NaN
df.replace("?", np.nan, inplace = True)
df.head()
```

```python
# Evaluating for Missing Data
missing_data = df.isnull()
missing_data.head()
```

- "True" stands for missing value, while "False" stands for not missing value.

```python
# Count missing values in each column
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

**Replacing Values of Numerical Variable**

```python
# Calculate the average of the column "normalized-losses"
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
# Replace "NaN" by mean value in "normalized-losses" column
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
# Calculate the mean value for 'bore' column
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
# Replace NaN by mean value
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
#Calculate the mean vaule for "stroke" column
avg_stroke = df["stroke"].astype("float").mean(axis = 0)
print("Average of stroke:", avg_stroke)
```

```python
# replace NaN by mean value in "stroke" column
df["stroke"].replace(np.nan, avg_stroke, inplace = True)
```

```python
# Calculate the mean value for the 'horsepower' column
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
```

```python
# Replace "NaN" by mean value
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
# Calculate the mean value for 'peak-rpm' column
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
```

```python
# Replace NaN by mean value
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

### Correcting Data Format

```python
# listing datatype of each columns
df.dtypes
```

- Numerical variables should have type 'float' or 'int', and variables with strings such as categories should have type 'object'.
- 'bore' and 'stroke' variables are numerical values that describe the engines, so we should expect them to be of the type 'float' or 'int'; however, they are shown as type 'object'. 

```python
# Convert data types to proper format
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```

### Data Normalization

- Normalization is the process of transforming values of several variables into a similar range. Typical normalizations include scaling the variable so the variable average is 0, scaling the variable so the variance is 1, or scaling variable so the variable values range from 0 to 1

- To demonstrate normalization, let's say we want to scale the columns "length", "width" and "height"

```python
# replace (original value) by (original value)/(maximum value)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
```

### Binning

- Binning is a process of transforming continuous numerical variables into discrete categorical 'bins', for grouped analysis.

```python
# Convert data to correct format
df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

<p>We would like 3 bins of equal size bandwidth so we use numpy's <code>linspace(start_value, end_value, numbers_generated</code> function.</p>
<p>Since we want to include the minimum value of horsepower we want to set start_value=min(df["horsepower"]).</p>
<p>Since we want to include the maximum value of horsepower we want to set end_value=max(df["horsepower"]).</p>
<p>Since we are building 3 bins of equal length, there should be 4 dividers, so numbers_generated=4.</p>


```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
# apply the function "cut" the determine what each value of "df['horsepower']" belongs to.
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
# Lets see the number of vehicles in each bin
df["horsepower-binned"].value_counts()
```

```python
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

### Bins visualization

```python
# draw historgram of attribute "horsepower" with bins = 3
plt.pyplot.hist(df["horsepower"], bins = 3)

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```
